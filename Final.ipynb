```python
#AI Game mode
def AI_funct(minim, maxim, win):
    player_wins = 0
    machine_wins = 0
    attempts=3
    number = randint(minim, maxim)
    dumb = 0
    while (player_wins < win) and (machine_wins <win):
        #If there is no winner, the game is not finished
        guess=input("Guess: ")        
        if guess.isnumeric():
            #If the input is numeric, the program will start making operations
            guess = int(guess)
            if number == guess:
                #If the random number is equal to player's guess:
                player_wins += 1
                attempts = 3
                number = randint(minim, maxim)
                print(f'Congratulations! Your points:{player_wins}')
            elif (number != guess) and (attempts > 1):
                #If the random number is not equal to player's guess:
                if number < guess:
                    print("The number is lower!")
                elif number > guess:
                    print("The number is higher!")
                attempts -= 1
                print(f"Come on, you still have a chance! attempts:{attempts}")
            else:
                #If the random number is not equal to player's guess and he is about to lose:
                machine_wins += 1
                attempts = 3
                number = randint(minim, maxim)
                print(f'You are not lucky today... AI points: {machine_wins} ')
                
        elif dumb <2:
            #If the input is not numeric and dumb counter has not strike 3:
            print('Check your neuro... input number i guess <3')
            dumb += 1
            
        else: 
            #If you introduced 3 wrong strings
            print('Error 404: Brain not found')
            attempts -= 1
            
    if player_wins == win:
        #If you win
        print('Congratulations! You have won!')
  
    else:
        #If you lose
        print('What a bad luck you had, you lost :(')  
        
'''All the code of this mode is an advanced version of the last one, so it won't be commented.'''

def PvP_funct(p1, p2, win):
    p1_wins = 0
    p2_wins = 0
    attempts=3
    dumb = 0
    
    while (p1_wins < win) and (p2_wins <win):
        
        number = getpass.getpass('Introduce the number: ')
        string = getpass.getpass('Introduce the data: ')
        print(f'Clue: {string}')
        
        while number.isnumeric():
            
            number = int(number)
            guess=input("Guess: ")
        
            if guess.isnumeric():
                guess = int(guess)
                if number == guess:
                    p1_wins += 1
                    attempts = 3
                    print(f'Congratulations {p1}! You have {p1_wins} points')
                    number =str(number)
                    break
                elif (number != guess) and (attempts > 1):
                    number =str(number)
                    attempts -= 1
                    print(f"Come on, you still have a chance! attempts: {attempts}")
                else:
                    p2_wins += 1
                    attempts = 3
                    print(f'Congratulations {p2}! You have {p2_wins} points')
                    number =str(number)
                    break
                
            elif dumb <2: 
                print('Check your neuro... input number i guess <3')
                number =str(number)
                dumb += 1
            
            else: 
                print('Error 404: Brain not found')
                number =str(number)
                attempts -= 1
        else:        
            print("Almost bro, but it is not trivial.")
            sys.exit()

    if player_wins == win:
        print(f'Congratulations! You have won! Wanna try again?')
  
    else:
        print(f'Congratulations! You have won! Wanna try again?')
run_cell(6)
```
